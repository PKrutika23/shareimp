# HeatMap

We cannot make heatmap of a string values, only integer values..

# Violin plot
A violin plot plays a similar role as box plotand whisker plot. It shows the distrubution of quantitative data across several levelsof one(or more) categorical variables such that those distrubutions can be compared.

seaborn.set(style="whitegrid")
f=seaborn.load_dataset("fmri")
seaborn.violinplot(x="",y="",data=f)

```python
data_P1=sns.load_dataset("tips")
data_P1
```

```python
sns.violinplot(x="day",y="total_bill",data=data_P1,linewidth=2,palette="Dark2")
plt.show()
```

```python
sns.violinplot(x="time",y="total_bill",data=data_P1,order=["Lunch","Dinner"],
               saturation=0.5,inner="quartile")
#inner=box/quartile/point/stick/None
plt.show()
```

```python
sns.violinplot(x="day",y="total_bill",data=data_P1,color="r",hue="sex",split=True)
plt.show()
```

```python
# Single violin plot
sns.violinplot(x="total_bill",data=data_P1,color="m")
# we could also write x=data_P1["total_bill"]
plt.show()
```

# Pair Plot
Graph of all the combinations of the numerical data. A pair plot is also known as a scatterplot, in which one variable in the same data row is matched with another variable's value (All variable paired with all the other variables)
df=sns.load_dataset('')
sns.pairplot(df)
plt.show( )

```python
# Note the numerical data in the tips dataframe is tip,size,total_bill (3)
sns.pairplot(data_P1)
plt.show( )
```

```python
sns.pairplot(data_P1,vars=["total_bill","tip"],hue="sex",
             hue_order=["Female","Male"],markers=["*","o"])
# vars=[] is important only those data will be displayed ie.total_bill,tip
plt.show( )
```

```python
sns.pairplot(data_P1,x_vars=["total_bill","tip"],y_vars=["tip","size"],hue="sex",palette="flare")
# vars=[] is important only those data will be displayed ie.total_bill,tip
plt.show( )
```

```python
sns.pairplot(data_P1,kind="kde",diag_kind="hist")
# kind= reg/scatter/hist/kde
plt.show( )
```

## Multiple plot (Facet - Grid)

```python
var=sns.load_dataset("tips")
var
# Note, that the tip is dependent variable.
# Note, total_bill,sex,smoker,day,time,size is the independent variable. 
```

```python
fg=sns.FacetGrid(var,col="sex",hue="day")
fg.map(plt.scatter,"total_bill","tip").add_legend()
plt.show()
```

```python
fg=sns.FacetGrid(var,col="day",hue="sex",palette="Accent")
fg.map(plt.bar,"total_bill","tip").add_legend()
plt.show()
```

# JointPlot
A join plot allows to study the relationship between 2 numeric variables. 
The central chart displays their correlation. It is usually a scatterplot, a hexbin plot, a 2D histogram or 2D density plot.

## Univariate Analysis

```python
sns.jointplot(x="tip",y="total_bill",data=var,kind="hex")
plt.show=()
```

```python
sns.jointplot(x="tip",y="total_bill",data=var,kind="reg")
plt.show=()
```
